```
from zipfile import ZipFile
import pdb
# --------- Extarct some files in Dataset --------- #
fileName = 'fruits_small'
classes = ['Apple Pink Lady', 'Avocado', 'Kiwi', 'Lemon', 'Mango', 'Banana', 'Peach', 'Pomegranate']
with ZipFile('fruits.zip', 'r') as zipObj:
    # Iterate over the file names
    for fileName in zipObj.namelist():
        pdb.set_trace()
        # Check filename endswith csv
        if fileName in classes:
            # Extract a single file from zip
            zipObj.extract(fileName, 'temp_csv')
```

```
# import pdb; pdb.set_trace()
from shutil import copytree
import os

def mkdir(p):
  if not os.path.exists(p):
      os.mkdir(p)

def link(src, dst):
  if not os.path.exists(dst):
      copytree(src, dst)
      # os.symlink(src, dst, target_is_directory=True) <-- not working (!)


# new_path = os.path.join( os.getcwd(), 'Fruits_data/fruits-360-small')
# new_path = '../MNIST_fashion/fruits_data/fruits-360-small'
# os.mkdir(new_path)

classes = ['Apple Pink Lady', 'Avocado', 'Kiwi', 'Lemon', 'Mango', 'Banana', 'Peach', 'Pomegranate']

train_path_from = os.path.abspath('../fruits-360/fruits-360_dataset/Training')
test_path_from = os.path.abspath('../fruits-360/fruits-360_dataset/Test')

mkdir('../fruits-360/fruits-360_reduced')

train_path_to = os.path.abspath('../fruits-360/fruits-360_reduced/Training')
test_path_to  = os.path.abspath('../fruits-360/fruits-360_reduced/Test')

mkdir(train_path_to)
mkdir(test_path_to )

for c in classes:
    link(train_path_from + '/' + c, train_path_to + '/' + c)
    link(test_path_from + '/' + c,  test_path_to + '/' + c)
```
