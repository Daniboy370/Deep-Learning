##Validate on video footage##

```
os.chdir(PATH_dir+'/yolov3')

input_video = 'MOV_0038.mp4'
cap = cv2.VideoCapture(input_video)
# cap.isOpened() # Ensure video is loaded (!)

# Recording Video
fps = cap.get(cv2.CAP_PROP_FPS)  # Find original frames rate
width, height = int(cap.get(3)), int(cap.get(4))
fcc = cv2.VideoWriter_fourcc('D', 'I', 'V', 'X')
# NOTICE ! I flipped (height, width) because the video is NOT rotated
vid_out = cv2.VideoWriter("video_output.avi", fcc, fps, (height, width) ) # <-- !!! flipped dimensions

# ----- Extracting desired footage segments ----- #
fr_tot = cap.get(cv2.CAP_PROP_FRAME_COUNT)
fr_i, fr_0_a, fr_end_a = 0, 0.01*fr_tot, 0.28*fr_tot # initialize indices

while fr_i < fr_tot:
    _, frame = cap.read()

    # ----- Desirable segment for detection ----- #
    if ( (fr_i > fr_0_a) & (fr_i < fr_end_a) ):
        frame = cv2.flip(cv2.transpose(frame), flipCode=1) # rotate 90 degrees
        # frame = cv2.flip(frame, -1)
        # -------- run YOLOv3 detection --------- #
        frame_BB = run( frame )
        vid_out.write( frame_BB )        # Recording Video

        # ---------- Print live index ----------- #    
        if fr_i % 5 == 0:
            print("Progress: {:.2f} [%]".format( 100*(fr_i - fr_0_a)/(fr_end_a - fr_0_a) ))
            # cv2_imshow( frame_BB )     # uncomment for online presentation

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    # else:
    #     break
    
    fr_i += 1

# -------- Close and Save video footage --------- #
cap.release()
vid_out.release()
cv2.destroyAllWindows()
```
